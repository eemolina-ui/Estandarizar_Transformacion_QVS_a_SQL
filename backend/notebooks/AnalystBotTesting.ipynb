```python
import sys
sys.path.append('../')
import os 
from services.DataAnalysisFlow.DataAnalyst import DataAnalysisFlow

from dotenv import load_dotenv
load_dotenv('../.env')

import boto3
import base64
import io
from PIL import Image

bedrock_client = boto3.client(service_name='bedrock-runtime')

AnalystBot = DataAnalysisFlow(
    llm_client=bedrock_client,
    bedrock_model_id=os.environ['CLAUDE_MODEL_ID'],
    data_directory='../services/DataAnalysisFlow/data/',
    forecast_api_url='http://localhost:8000/forecast/'
    )

print(AnalystBot.available_datasets)

def display_response(response):
    if 'text' in response:
        print('Respuesta en texto:', response['text'])
    
    if 'data' in response:
        print('Datos:')
        display(response['data'])
    
    if 'images' in response and response['images']:
        print('Gráfica:')
        for image in response['images']:
            image_bytes = base64.b64decode(image.split(',')[1])
            img = Image.open(io.BytesIO(image_bytes))
            display(img)
    if 'error' in response:
        print('Error:', response['error'])

    if 'plan' in response:
        print('Plan:', response['plan'])

    if 'generated_code' in response:
        print('Código generado:', response['generated_code'])
```
